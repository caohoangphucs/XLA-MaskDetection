```python
import splitfolders
import os

input_folder = './MaskDetection/data'
output_folder = './MaskDetection/split_data'

# Clear the output folder if it already exists to ensure a clean split
if os.path.exists(output_folder):
    import shutil
    shutil.rmtree(output_folder)


# Split
splitfolders.ratio(input_folder, output=output_folder, seed=42, ratio=(0.8, 0.1, 0.1), group_prefix=None)
#Clear raw data
shutil.rmtree(input_folder)
print(f"Dataset split into train, val, and test folders in '{output_folder}' with ratio (80%, 10%, 10%).")
```

Now resize image

```python
from PIL import Image
import os

input_dir = "/content/MaskDetection/split_data"
output_dir = "/content/MaskDetection/resized_data"
size = (224, 224)

os.makedirs(output_dir, exist_ok=True)

for root, dirs, files in os.walk(input_dir):
    relative_path = os.path.relpath(root, input_dir)
    save_path = os.path.join(output_dir, relative_path)
    os.makedirs(save_path, exist_ok=True)

    for file in files:
        if file.lower().endswith(('.jpg', '.png', '.jpeg')):
            img_path = os.path.join(root, file)
            img = Image.open(img_path).convert("RGB")
            img = img.resize(size)
            img.save(os.path.join(save_path, file))

print("Done!")
```
